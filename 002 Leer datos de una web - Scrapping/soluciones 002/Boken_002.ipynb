Por cada elemento `th` de la tabla vamos a leer el nombre de las columnas

Por cada elemento `tr` que son los rows de la tabla, PERO filtrando por los que tienen el class `10_2020` tal y como se ve en el html

```html
<tr class="10_2020 GGAL trOrder accion-Call acciones-20 acciones-10">
```

Ya que hay otros rows cuyo contenido, que no queremos.

```html
<tr class="table-mobile-tr ">
```

Hay varias columnas, cuyo contenido esta dentro de otros labels `strong` `small` y por eso tenemos que tratar de obtenerlo y si falla `try: except`, cogemos el contenido genérico.

````python
rows = table.find_all('tr', class_='10_2020')
data = []
for row in rows:
    columns = []
    for column in row.find_all('td'):
        try:
            columns.append([v.contents[0] for v in column.find_all('small')][0])
        except:
            try:
                columns.append([v.contents[0] for v in column.find_all('strong')][0])
            except:
                columns.append(column.contents[0].replace('\n','').replace('\r',''))
            
    columns2 = [str(c).replace(' ','') for c in columns[:13]]
    columns2.append(columns[13])
    data.append(columns2)
    
data
````

#### Ahora que ya lo tenemos montamos el DataFrame con el indice correcto

````python
df = pd.DataFrame(data, columns=headers)
df
````
