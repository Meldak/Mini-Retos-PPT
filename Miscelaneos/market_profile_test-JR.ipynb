```python
from market_profile import MarketProfile
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.pylab as pylb

%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv('^DJI.csv', parse_dates=True, index_col='Date',
                usecols=['Date', 'Open', 'High',
                         'Low', 'Close','Volume']).dropna()
```

```python
df = df.tail(900)
```

```python
len(df)
```

```python
from market_profile import MarketProfile

i = 25 # Dias a calcular POC
poc_values = []
mp = MarketProfile(df, tick_size=1)

for i in range(len(df)) :   
    poc = mp[df.index[i-25]:df.index[i]].poc_price
    i = i+1
    poc_values.append(poc)
    
df['POC'] = np.NaN
df.iloc[(len(df) - len(poc_values)):,-1:] = poc_values

df = df.dropna()
```
